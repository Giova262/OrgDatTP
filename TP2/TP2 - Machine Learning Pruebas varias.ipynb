# Random Forest Hiperparametros

```python
d1 = {
    
    'max_feactures': [9,15,18,19,20,25,38,21,22], 
    'error': [0.369,0.3566,0.35551,0.35537,0.35549,0.35621,0.46208,0.35513,0.35518]
}
df1 = pd.DataFrame(data=d1)
```

```python
#Random Forest max_feactures-error max_features vs Error

f, ax = plt.subplots(figsize=(10.0, 10.0))
sns.despine(f, left=True, bottom=True)
sns.lineplot(x="max_feactures", y="error", data=df1)
plt.xlim(17, 23)
plt.ylim(0.3540, 0.3560)
```

```python
f, ax = plt.subplots(figsize=(10.0, 10.0))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="max_feactures", y="error", data=df1)
plt.xlim(17, 25)
plt.ylim(0.3540, 0.3560)
```
