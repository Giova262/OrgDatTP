```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from matplotlib.pyplot import *
sns.set(style="darkgrid")
```

```python
#Abro el csv
df_zonap = pd.read_csv('../Archivos CSV/trainTP12019.csv'
            ,dtype={
             "id":np.int32
            ,"titulo":'category'
            ,"fecha":str 
            ,"ciudad": 'category'
            ,"provincia": 'category'
            ,"tipodepropiedad": 'category' })

df_zonap.head(3)
```

```python
#Cambio la fecha a Datatime
df_zonap['fecha'] = pd.to_datetime(df_zonap['fecha'])
```

```python
#Creo formato para el heatmap
new = df_zonap[['fecha', 'precio']].copy()
new['year'] = new.fecha.dt.year
new['month'] = new.fecha.dt.month
del new['fecha']
new.head()
```

```python
import numpy as np
import seaborn as sns
sns.set(style="ticks")

x = df_zonap.fecha.dt.dayofweek
y = df_zonap.precio

sns.jointplot(x, y, kind="hex", color="#4CB391")
```

```python
#Precios a lo largo del tiempo (Precios fueron aumentando inflacion)
fig, ax = pyplot.subplots(figsize=(20, 20))
heatmap1_data = pd.pivot_table(new, values='precio', 
                     index=['month'], 
                     columns='year')
sns.heatmap(heatmap1_data, cmap="YlGnBu", linewidths=.5)
```

```python
#Caracteristicas de cada tipo de propiedad
labels=np.array(['usosmultiples', 'habitaciones', 'garages', 'banos', 'gimnasio', 'piscina'])
stats=df_zonap.loc[2,labels].values

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title([df_zonap.loc[1,"tipodepropiedad"]])
ax.grid(True)
```

```python
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```
